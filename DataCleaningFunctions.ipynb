```python
import pandas as pd

def type_check_data(df: pd.DataFrame):
    if not isinstance(df, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame.")
```

```python
from typing import List
import pandas as pd

def print_unique_values(column_names: List[str], data: pd.DataFrame):
    type_check_data(data)
    for columns in column_names:
        print(f"{columns} Column: {data[columns].unique()}")
```

```python
import pandas as pd

def replace_data_values(df: pd.DataFrame, column_name: str, to_replace, value):
    type_check_data(df)
    df[column_name].replace(to_replace, value, inplace=True)
```

```python
from typing import List
import pandas as pd

def check_numerical_column_errors(df: pd.DataFrame, column_names: List[str]):
    type_check_data(df)
    for column in column_names:
        df[column] = pd.to_numeric(df[column], errors='coerce')
```
